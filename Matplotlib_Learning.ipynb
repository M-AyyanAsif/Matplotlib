```python
import matplotlib.pyplot as plt
hours_studied=[1,2,3,4,5,6,7,8,9]
marks_gain=[44,76,67,98,34,76,67,55,87]
plt.scatter(hours_studied,marks_gain,color="red",marker="o",label="Student_data")
plt.legend(fontsize=10)
plt.grid(True)
plt.xlabel("No.of_Hours")
plt.ylabel("Marks_Gain By Students")
plt.title("Numers Gain By Studied Hours")
plt.show()
```

```python
# Subplots
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[10,22,45,66,33,61,70]

plt.subplot(1,2,1)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Line Chart")
plt.plot(x,y)

plt.subplot(1,2,2)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Bar Chart")
plt.bar(x,y)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig ,ax= plt.subplots(1,2, figsize=(10,7))
x=[1,2,3,4,5]
y=[22,43,26,37,114]

ax[0].plot(x,y, color="red", linestyle="--",linewidth=7)
ax[0].set_title("Line Graph")

ax[1].bar(x,y,color="yellow")
ax[1].set_title("Line Graph")
plt.tight_layout()
plt.show()
```
