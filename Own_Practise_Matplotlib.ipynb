```python
import matplotlib.pyplot as plt
```

```python
Sale=[1000,780,820,520,1020]
Products=["Shampoo","Soap","Fruits","Banana","Balls"]
plt.plot(Products,Sale, linestyle="--", marker="o",color="Green",linewidth=3, label="Products Sales by Qtr")
plt.xlabel("Total_Sales")
plt.ylabel("Sale")
plt.xlim(1,5)
plt.ylim(0,1000)
plt.title("Products sales by Quaters")
plt.legend(fontsize=11)
plt.grid(linestyle="--",color="black",linewidth=2)
plt.show()
```

```python
Products=["Shampoo","Soap","Fruits","Banana","Balls"]
Sale=[1000,780,820,520,1020]
plt.bar(Products,Sale, color="Red", label="Sales by Week")
plt.xlabel("All Weeks")
plt.ylabel("Sales by Week")
plt.legend(fontsize=9)
plt.xticks(["Shampoo","Soap","Fruits","Banana","Balls"],["W1","W2","W3","W4","W5"])
plt.title("Sales of Products In a Month")
plt.show()
```

```python
import matplotlib.pyplot as plt
num=[11,18,22,33,44,55,57,59,66,61,77,88,99]
plt.hist(num,bins=9,color="green",edgecolor="black",label="Catogery of numbers")
plt.xlabel("Categories")
plt.ylabel("Numbers scored by students")
plt.title("Numers gained by each students")
plt.legend(fontsize=8)
plt.show()
```

```python
import matplotlib.pyplot as plt
numbers1=[23,37,45,54,61,76,85,99]
numbers2=[29,39,41,59,68,71,82,92]

hours1=[2,4,5,7,8,2,4,1]
hours2=[5,7,4,9,8,7,2,2]

plt.scatter(numbers1,hours1,color="blue",marker="o",label="Students_1 Record")
plt.scatter(numbers2,hours2,color="yellow",marker="o",label="Students_2 Record")
plt.xlabel("Numbers of Section A B")
plt.ylabel("Numbers of Hours Studied BY A B")
plt.title("Numbers gained by hours")
plt.legend(fontsize=8)
plt.grid(True)
plt.show()
```

```python
#subplots
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[11,23,8,44,33,60,76]

plt.subplot(1,2,1)
plt.plot(x,y, color="blue",linewidth=3)
plt.title("Line Graph")

plt.subplot(1,2,2)
plt.bar(x,y, color="yellow",linewidth=3)
plt.title("Bar Graph")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,2, figsize=(7,5))

x=[1,2,3,4,5,6,7]
y=[23,11,45,33,56,66,71]
ax[0].plot(x,y, color="green")
ax[0].set_title("Line Graph")

ax[1].bar(x,y, color="red")
ax[1].set_title("Bar Graph")

fig.suptitle("Comparison of Bar and Line Chart")
plt.tight_layout()
plt.show()
```
